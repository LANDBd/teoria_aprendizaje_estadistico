```python
def es_primo(n):

  if(n==1):
    return False

  for i in range(2,n):
    if(n % i == 0):
      return False
  
  return True
```

```python
es_primo(29996224275833)
```

```python
def es_primo_rapido(n):

  if(n==1):
    return False

  for i in range(2,int(sqrt(n) + 1)):
    if(n % i == 0):
      return False

  return True
```

```python
es_primo_rapido(54)
```

# Recursividad

```python
def factorial(n):
    if n < 0:
        raise ValueError("El número debe ser no negativo")
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
factorial(10000000)
```

```python
def fibonacci(n):
  # Esta función regresa el n-ésimo término de Fibonacci

  if(n==1):
    return 0
  elif(n==2):
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(7)
```

```python
import time
```

```python
inicio = time.time()
numero = fibonacci(40)
fin = time.time()

print(f'El termino de fibonacci calculado es {numero}')
print(fin - inicio)
```

# Algo más complicado

```python
def burbuja(lista):
    n = len(lista)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lista[j] > lista[j + 1]:
                # Intercambiar los elementos si están en el orden incorrecto
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
    return lista

# Ejemplo de uso
numeros = [5, 3, 8, 4, 2]
print("Lista original:", numeros)
ordenada = burbuja(numeros.copy())
print("Lista ordenada:", ordenada)
```

### Atributos e instancias de una clase

Una primera clase

```python
class CuentaBancaria:
    def __init__(self, titular='Sixto', saldo=0):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, cantidad):
        self.saldo += cantidad

    def retirar(self, cantidad):
        if cantidad <= self.saldo:
            self.saldo -= cantidad
        else:
            print("Fondos insuficientes.")

    def mostrar_saldo(self):
        return f"Saldo de {self.titular}: {self.saldo}"
```

```python
# Uso
cuenta = CuentaBancaria('David', 1000000)
```

```python
cuenta.saldo
```

```python
cuenta.depositar(1000)
```

```python
cuenta.saldo
```

```python
cuenta.retirar(5000)
print(cuenta.mostrar_saldo())
```

```python
class SecuenciaGenomica:
    def __init__(self, cadena):
        self.cadena = cadena

    def longitud(self):
        return len(self.cadena)
    
    def complemento(self):
        secuencia_complemento = ""

        for nucleotido in self.cadena:
            if(nucleotido == "A"):
                secuencia_complemento += "T"
            elif(nucleotido == "T"):
                secuencia_complemento += "A"

            # Modificar función para que dé el complemento correcto
            else:
                secuencia_complemento += "0"
        return secuencia_complemento
                
cadena = SecuenciaGenomica("ATCGGATTA")
print(cadena.complemento())
```

Método de clase

```python
class Carro:
    # Atributo de clase
    marca = "Toyota"

    def __init__(self, modelo, color):
        self.modelo = modelo
        self.color = color

    @classmethod
    def cambiar_marca(cls, nueva_marca):
        cls.marca = nueva_marca
        print(f"Marca cambiada a {cls.marca}")
```

```python
# Crear una instancia
mi_carro = Carro("Fusion", "Rojo")
```

```python
# Llamar al método de clase para cambiar la marca
Carro.cambiar_marca("Ford")

# Verificar el cambio de marca
print(f"La marca del carro es ahora: {mi_carro.marca}")
```

```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Método para la representación del objeto

    # Como representar el objeto con la funcion print
    def __repr__(self):
        return f"Punto({self.x}, {self.y})"

    # Método para sumar puntos
    # Como operar con el signo "+" dos puntos
    def __add__(self, otro_punto):
        return Punto(self.x + otro_punto.x, self.y + otro_punto.y)
```

```python
# Crear dos instancias de Punto
punto1 = Punto(2, 3)
punto2 = Punto(5, 7)

# Sumar los puntos
resultado = punto1 + punto2
print(resultado)  
```

Otra librería: scipy

```python
from scipy import linalg
import numpy as np

# Resolver Ax = b
A = np.array([[3, 2], [1, 4]])
b = np.array([6, 5])
x = linalg.solve(A, b)

print(f"Solución: x = {x}")
```

```python
from scipy.optimize import minimize

def f(x):
    return x[0]**2 + x[1]**2 + 3

res = minimize(f, x0=[1, 1])
print(f"Mínimo encontrado en: {res.x}, valor mínimo: {res.fun}")
```

```python
from scipy import integrate

# Integrar f(x) = x^2 de 0 a 1
result, error = integrate.quad(lambda x: x**2, 0, 1)

print(f"Resultado de la integral: {result}, Error estimado: {error}")
```
